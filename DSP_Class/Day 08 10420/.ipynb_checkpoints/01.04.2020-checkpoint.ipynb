```python
def closed_value(a, n):
    d = np.abs(a-n)
    idx = np.argmin(d)
    i,j = idx//a.shape[1],idx%a.shape[1]
    return i,j,a[i,j]
```

```python
closed_value(a2,2)
```

```python
a2.shape
```

```python
d = np.abs(a2-2)
np.argsort(d,axis=None)
```

```python
def closed_value_extended(a, n, k):
    d = np.abs(a-n)
    idx = np.argsort(d,axis=None)[:k]
    l = []
    for ix in idx:
        i,j = ix//a.shape[1],ix%a.shape[1]
        l.append((i,j,a[i,j]))
    return l
```

```python
closed_value_extended(a2, -2, 5)
```

```python
def from_txt_to_iris_class(x):
    if x == b'Iris-setosa': 
        return 1
    elif x == b'Iris-versicolor': 
        return 2
    elif x == b'Iris-virginica': 
        return 3
    else:
        return np.nan
    
a = np.loadtxt('iris.csv', skiprows=1, 
               delimiter=',', 
               converters ={4: from_txt_to_iris_class})
a
```

```python
def solve_for_x(A, z):
    """ Solve the equation Ax = z and return the vector x. """
    """ A(3,3), z(3,1) ==> x(3,1)"""
    return np.matmul(np.linalg.inv(A), z)
```

```python
np.linalg.inv(A)
```

```python
solve_for_x(A,z)
```

```python
def print_rows(matrix):
    """ Print the rows of the matrix, one per line. """
    # Reminder: matrix.shape[0] contains the number of rows
    for i in range(matrix.shape[0]):
        print(matrix[i])

def print_cols(matrix):
    """ Print the columns of the matrix, one per line. """
    # Reminder: matrix.shape[1] contains the number of columns
    for i in range(matrix.shape[1]):
        print(matrix[:,i].T)
```

```python
print(A)
print('rows')
print_rows(A)
print('cols')
print_cols(A)
```

### Squared error 
$\sum_{x_i \in X} (x_i \times w^T - y_i)^2$

$X \in R^{n\times m} \rightarrow x_i \in R^{1 \times m}$

$w \in R^{m}$

$y \in R^{n} \rightarrow y_i \in R$

```python
def squared_error(X, w, y):
    """ Return the sum squared error for the the provided linear fit. 

    returns: 
      $\sum_{x_i \in X} (x_i \times w^T - y_i)^2$
     
    Arguments: 
       X - An n x m numpy array, one data point per row. 
       w - A length m numpy array containing weights. 
       y - a length n numpy array containing target values. 

    Returns: 
      A single number representing the total squared error. 

    >>> print squared_error(np.array([[-1.0, 1.0, 0], [-2.0, 2.0, 3.0]]),\
                            np.array([1, 1, 1]), np.array([1, 2]))
    2.0

    """
    s = 0
    for i,xi in enumerate(X):
        s += (np.matmul(xi,w.T) - y[i])**2
    return s

squared_error(np.array([[-1.0, 1.0, 0], [-2.0, 2.0, 3.0]]),\
                            np.array([1, 1, 1]), np.array([1, 2]))
```

```python
def squared_error_fast(X, w, y):
    return np.sum((np.matmul(X,w.T) - y.T)**2)  # n x m * m x 1 => n x 1

squared_error_fast(np.array([[-1.0, 1.0, 0], [-2.0, 2.0, 3.0]]),\
                            np.array([1, 1, 1]), np.array([1, 2]))
```
