```python
df_epl = pd.read_csv('epldata_final.csv')
df_epl.head()
```

```python
# tổng giá trị cầu thủ theo câu lạc bộ
df_epl[['club', 'market_value']].groupby('club').agg('sum')
```

```python
# Câu lạc bộ có giá trị cầu thủ lớn nhất
dfg = df_epl[['club', 'market_value']].groupby('club').agg('sum')
dfg.idxmax()
```

```python
dfg.max()
```

```python
df_epl[['position','market_value']].groupby('position').max() #.loc['GK']
```

```python
# Thủ môn đắt nhất (lấy 1 người)
df_gk = df_epl[['name','market_value',]][df_epl.position=='GK']
df_epl.iloc[df_gk['market_value'].idxmax()]
```

```python
# Lấy tất cả thủ môn đắt nhất
# Cách 1
df_gk[df_gk.market_value == df_gk['market_value'].max()]
```

```python
# Cách 2
df_gk = df_epl[['name', 'market_value']][df_epl.position == 'GK']
df_gk.sort_values(by=['market_value']).tail(2)
```

```python
# Cách 3
df_gk.where(df_gk.market_value==df_gk.market_value.max()).dropna()
```

```python
# Quốc gia có ít cầu thủ nhất
ct = df_epl[['name','nationality']].groupby('nationality').count()
ct.sort_values(by=['name'])
```

```python
ct[ct.name == ct.name.min()]
```

```python
# Quốc gia có nhiều cầu thủ nhất
ct[ct.name == ct.name.max()]
```

```python
# chua dung
df_epl[df_epl.market_value == df_epl.market_value.max()]
```
