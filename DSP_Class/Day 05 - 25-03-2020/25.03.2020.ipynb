### Generator

```python
def myreadline(fn):
    f = open(fn, 'r')
    for l in f.readlines():
        yield l
    f.close()
```

```python
f = myreadline('word_count_data.txt')
next(f)
```

### Recursive function

```python
l1 = [2,5,3,7,3,7,4]
l2 = [2,4,5.6, 'ste', 5.9, 8, [23,5,9, [(3,56, (3,4.7,8),23),5,8,2]], '32',5, (332,'332',4),'sdsd']
```

```python
def sumof(l):
    if len(l) == 1:
        return l[0]
    elif len(l) == 0:
        return 0
    else:
        #return sum of element in the pozition 0 + the string from the position 1
        return l[0] + sumof(l[1:])
```

```python
sumof(l1)
```

```python
def sumof2(l):
    #if the length of l is 1
    if len(l) == 1:
        #and it is instance of int or float
        if isinstance(l[0],(int, float)):
            #return the value of l
            return l[0]
        #if it is a string, return 0
        elif type(l[0]) == str:
            return 0
        #if it is a list or tuple
        elif isinstance(l[0],(list, tuple)):
            #count the sum of l[0]
            return sumof2(l[0])
    #if the length = 0
    elif len(l) == 0:
        return 0
    #if the length more than 1
    else:
        #if l[0] is int / float: sum = sum of ( the value first + call recursively function sumof2 for the next element)
        if isinstance(l[0], (int, float)):
            return l[0] + sumof2(l[1:])
        #if l[0] is string:
        elif type(l[0]) == str:
            #chỉ tính sumof2 tiếp cho string còn lại
            return sumof2(l[1:])
        #nếu là list hay tuple, lại tính 
        elif isinstance(l[0], (list, tuple)):
            return sumof2(l[0]) + sumof2(l[1:])
```

```python
sumof2(l2)
```

```python
def show_folder(path, indent):
    ls = os.listdir(path)
    for l in ls:
        if os.path.isfile(os.path.join(path,l)):
            print(indent + '- ' + l)
        elif os.path.isdir(os.path.join(path,l)):
            print(indent + '* ' + l)
            show_folder(os.path.join(path,l), indent + indent*4)
```
