#### Word count by using map and reduce function

```python
f = open('word_count_data.txt', 'r')
# print(f.readline())
# for l in f:
#     print(l)
sl = f.read().split()
f.close()
sl
```

```python
def myfunc(s):
    ch = '!@#$%^&*(){}[],.<>-'
    for c in ch:
        s = s.replace(c,'')
    if len(s) == 1 and s != 'a':
        return 0
    elif len(s) > 1:
        return 1
    else:
        return 0
```

```python
sl1 = list(map(myfunc, sl))
sl1
```

```python
from functools import reduce
t = reduce(lambda x, y: x+y, sl1)
t
```

```python
f = open('word_count_data.txt', 'r')
# print(f.readline())
# for l in f:
#     print(l)
sl = f.readlines()
f.close()
sl
```

```python
sl2 = list(map(lambda x:  x.split(),sl))
sl2
```

```python
a = [list(map(lambda x: 1, s)) for s in map(lambda x:  x.split(),sl)]
a
```

```python
a1 = [reduce(lambda x,y: x+y, i) for i in a if len(i) > 0]
a1
```

```python
reduce(lambda x,y: x+y, a1)
```

### Generator

```python
def myreadline(fn):
    f = open(fn, 'r')
    for l in f.readlines():
        yield l
    f.close()
```

```python
f = myreadline('word_count_data.txt')
next(f)
```

```python
next(f)
```

```python
for l in f:
    print(l)
```

### Recursive function

```python
l1 = [2,5,3,7,3,7,4]
l2 = [2,4,5.6, 'ste', 5.9, 8, [23,5,9, [(3,56, (3,4.7,8),23),5,8,2]], '32',5, (332,'332',4),'sdsd']
```

```python
def sumof(l):
    if len(l) == 1:
        return l[0]
    elif len(l) == 0:
        return 0
    else:
        return l[0] + sumof(l[1:])
```

```python
sumof(l1)
```

```python
def sumof2(l):
    if len(l) == 1:
        if isinstance(l[0],(int, float)):
            return l[0]
        elif type(l[0]) == str:
            return 0
        elif isinstance(l[0],(list, tuple)):
            return sumof2(l[0])
    elif len(l) == 0:
        return 0
    else:
        if isinstance(l[0], (int, float)):
            return l[0] + sumof2(l[1:])
        elif type(l[0]) == str:
            return sumof2(l[1:])
        elif isinstance(l[0], (list, tuple)):
            return sumof2(l[0]) + sumof2(l[1:])
```

```python
sumof2(l2)
```

```python
def show_folder(path, indent):
    ls = os.listdir(path)
    for l in ls:
        if os.path.isfile(os.path.join(path,l)):
            print(indent + '- ' + l)
        elif os.path.isdir(os.path.join(path,l)):
            print(indent + '+ ' + l)
            show_folder(os.path.join(path,l), indent + ' '*4)
```

```python
show_folder('/Vicohub/Python','')
```
